```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
fig,ax = plt.subplots(figsize = (16,10))

Pegasus_loss = pd.read_csv('/content/history_pegasus.csv', usecols=['Step', 'Value'])
plt.plot(Pegasus_loss.Step, Pegasus_loss.Value, lw=1.5, label='Pegasus_loss')
Bart_loss = pd.read_csv('/content/history_bart.csv', usecols=['Step', 'Value'])
plt.plot(Bart_loss.Step, Bart_loss.Value, lw=1.5, label='Bart_loss')
Gpt2_loss = pd.read_csv('/content/history_GPT2.csv', usecols=['Step', 'Value'])
plt.plot(Gpt2_loss .Step, Gpt2_loss .Value, lw=1.5, label='Gpt2_loss')
T5_loss= pd.read_csv('/content/history_t5.csv', usecols=['Step', 'Value'])
plt.plot(T5_loss.Step, T5_loss.Value, lw=1.5, label='T5_loss',color='purple')

plt.grid(color='0.9')
plt.ylabel("Value")
plt.xlabel("Step")
plt.legend(loc=0)
plt.show()
```
